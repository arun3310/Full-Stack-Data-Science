```python
data = pd.read_csv('titanic.csv')
data
```

```python
data['Survived'].value_counts().values
```

```python
plt.figure(figsize=(14, 10))

labels = ['1st class','2nd class','3rd class']
explode = [0,.1,.1]
colors = ['Red','Green','Blue']
x = data['Pclass'].value_counts().values

plt.title("Proporation of Pclass ", size = 9)
plt.pie(x,explode,labels,colors,autopct = '%1.1f%%')
```

```python
plt.figure(figsize=(14, 10))

labels = ['1st class','2nd class','3rd class']
explode = [0,.1,.1]
colors = ['Red','Green','Blue']
x = data['Pclass'].value_counts().values

plt.title("Proporation of pclass ", size = 9)
plt.pie(x,explode,labels,colors,autopct = '%1.1f%%')
```
